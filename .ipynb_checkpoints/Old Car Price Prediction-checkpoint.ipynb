```python
cat_cols = ["seller_type", "manufacturer", "model"]
```

```python
dummies = pd.get_dummies(df[cat_cols], drop_first=True, dtype=int)
```

```python
data = pd.concat([df.drop(columns=cat_cols), dummies], axis=1)
```

```python
num_cols=["year", "selling_price","km_driven","mileage","engine","max_power"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
data[num_cols]=scaler.fit_transform(data[num_cols])
```

```python
X = data.drop("selling_price", axis=1)
y = data["selling_price"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=234)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
lr = LinearRegression()
rf = RandomForestRegressor(n_estimators=200)
dt = DecisionTreeRegressor()
```

```python
lr.fit(X_tr, y_tr)
rf.fit(X_tr, y_tr)
dt.fit(X_tr, y_tr)
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score
```

```python
lr_pred = lr.predict(X_te)
rf_pred = rf.predict(X_te)
dt_pred = dt.predict(X_te)
```

```python
lr_rmse = root_mean_squared_error(y_te, lr_pred)
rf_rmse = root_mean_squared_error(y_te, rf_pred)
dt_rmse = root_mean_squared_error(y_te, dt_pred)
```

```python
lr_r2 = r2_score(y_te, lr_pred)
rf_r2 = r2_score(y_te, rf_pred)
dt_r2 = r2_score(y_te, dt_pred)
```

```python
print(f"Linear Regression - RMSE {lr_rmse} and R2 Score {lr_r2}")
print(f"Random Forest - RMSE {rf_rmse} and R2 Score {rf_r2}")
print(f"Decision Tree - RMSE {dt_rmse} and R2 Score {dt_r2}")
```
