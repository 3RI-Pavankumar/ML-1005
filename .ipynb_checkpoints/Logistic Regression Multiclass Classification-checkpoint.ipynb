```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=['sep_len', 'sep_wid', 'pet_len', 'pet_wid'])
df 
```

```python
df["target"]=iris.target
```

```python
df.head()
```

```python
df["target"] = list(map(lambda x:"Setosa" if x==0 else ("Versicolor" if x==1 else "Verginica"), iris.target))
```

```python
df.head()
```

```python
sns.scatterplot(df, x="sep_len", y="sep_wid", hue="target", palette="rocket")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.kdeplot(df, x="sep_len", y="sep_wid", hue="target", palette="rocket")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.scatterplot(df, x="pet_len", y="pet_wid", hue="target", palette="rocket")
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
sns.kdeplot(df, x="pet_len", y="pet_wid", hue="target", palette="rocket")
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
sns.pairplot(df, hue="target", palette="rocket")
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, train_size=0.75, random_state=523)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_tr, y_tr)
```

```python
model.score(X_tr, y_tr)
```

```python
y_pr = model.predict(X_te)
```

```python
accuracy_score(y_te, y_pr)
```

```python
confusion_matrix(y_te, y_pr)
```

```python
sns.heatmap(confusion_matrix(y_te, y_pr), annot=True)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
print(classification_report(y_te, y_pr))
```
