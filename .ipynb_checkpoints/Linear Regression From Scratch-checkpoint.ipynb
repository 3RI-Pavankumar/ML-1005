```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
Area = np.array([350, 430, 455, 557, 640, 700, 770, 860, 900])
Price = np.array([11, 22, 41, 38, 57, 74, 70, 79, 88])
```

```python
Area
```

```python
Price
```

```python
plt.figure(figsize=(10, 5))
plt.xlabel("Area in Square Feet")
plt.ylabel("Price in Lakh")
plt.title("Area vs Price")
plt.scatter(Area, Price, marker="*", color="red")
plt.xticks(range(300, 1000, 100))
plt.grid(True)
plt.show()
```

```python
area_mean = Area.mean()
price_mean = Price.mean()
area_mean, price_mean
```

```python
coef = np.sum((Area-area_mean)*(Price-price_mean))/np.sum(np.square(Area-area_mean))
coef
```

```python
intercept = price_mean - area_mean * coef
intercept
```

```python
# y = mx + c
# y = coef * area + intercept
y_pred = coef * Area + intercept
y_pred
```

```python
plt.figure(figsize=(10, 5))
plt.xlabel("Area in Square Feet")
plt.ylabel("Price in Lakh")
plt.title("Area vs Price")
plt.scatter(Area, Price, marker="*", color="red")
plt.plot(Area, Price, color="red")
plt.scatter(Area, y_pred, marker=".", color="green")
plt.plot(Area, y_pred, color="green")
plt.xticks(range(300, 1000, 100))
plt.grid(True)
plt.show()
```

```python
MSE = np.sum(np.square(Price - y_pred))/len(Price)
MSE
```

```python
MAE = np.sum(np.abs(Price - y_pred))/len(Price)
MAE
```

```python
SSR = np.sum(np.square(Price - y_pred))
SSR
```

```python
Pred_price = np.round(y_pred, decimals=2)
```

```python
Act_price = Price
```

```python
Pred_price
```

```python
Act_price
```

```python
r2 = 1-(np.sum(np.square(Act_price-Pred_price)) / np.sum(np.square(Act_price-Act_price.mean())))
```
