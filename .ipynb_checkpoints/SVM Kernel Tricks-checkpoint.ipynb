```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.datasets import make_circles
from sklearn.model_selection import GridSearchCV
```

```python
X, y = make_circles(n_samples=1000, factor=0.6, noise=0.05)
```

```python
X
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=54)
```

```python
m_lin = SVC(kernel="linear")
```

```python
m_lin.fit(X_tr, y_tr)
print("Training Score - Linear Kernel :",m_lin.score(X_tr, y_tr))
print("Testing Score - Linear Kernel :",accuracy_score(y_te, m_lin.predict(X_te)))
```

```python
m_poly = SVC(kernel="poly", degree=2, C=2)
```

```python
m_poly.fit(X_tr, y_tr)
print("Training Score - Polynomial Kernel :",m_poly.score(X_tr, y_tr))
print("Testing Score - Polynomial Kernel :",accuracy_score(y_te, m_poly.predict(X_te)))
```

```python
X1, y1 = make_circles(n_samples=1000, factor=0.6, noise=0.3)
```

```python
plt.scatter(X1[y1==0, 0], X1[y1==0, 1])
plt.scatter(X1[y1==1, 0], X1[y1==1, 1])
plt.show()
```

```python
X_tr1, X_te1, y_tr1, y_te1 = train_test_split(X1, y1, test_size=0.3, random_state=54)
```

```python
m_poly2 = SVC(kernel="poly", degree=2, C=4)
```

```python
m_poly2.fit(X_tr1, y_tr1)
print("Training Score - Polynomial Kernel :",m_poly2.score(X_tr1, y_tr1))
print("Testing Score - Polynomial Kernel :",accuracy_score(y_te1, m_poly2.predict(X_te1)))
```

```python
m_rbf = SVC(kernel="rbf", gamma=0.3, degree=3, C=5)
```

```python
m_rbf.fit(X_tr1, y_tr1)
print("Training Score - RBF Kernel :",m_rbf.score(X_tr1, y_tr1))
print("Testing Score - RBF Kernel :",accuracy_score(y_te1, m_rbf.predict(X_te1)))
```

```python
params = {
    "C":np.arange(0.1, 3, 0.5),
    "kernel":['linear', 'poly', 'rbf', 'sigmoid'],
    "degree":range(1, 5),
    "gamma":['scale', 'auto', 1.4, 2.4, 3.4, 0.5, 0.9]
}
```

```python
gscv = GridSearchCV(estimator=SVC(), param_grid=params, n_jobs=-1, verbose=True, cv=10)
```

```python
gscv.fit(X_tr1, y_tr1)
```
