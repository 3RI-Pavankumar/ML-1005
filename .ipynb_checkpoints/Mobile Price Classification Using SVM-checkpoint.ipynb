```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, cross_val_score
```

```python
cols
```

```python
train[['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt',]]
```

```python
train[['n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi', 'price_range']]
```

```python
categorical = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi', 'price_range']
```

```python
numerical = [i for i in cols if i not in categorical]
```

```python
numerical
```

```python
train[numerical]
```

```python
scaler = StandardScaler()
```

```python
train[numerical]=scaler.fit_transform(train[numerical])
```

```python
train[numerical]
```

```python
train.head()
```

```python
X = train.drop("price_range", axis=1)
y = train["price_range"]
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=53)
```

```python
model=SVC(kernel="linear")
```

```python
model.fit(X_tr, y_tr)
```

```python
model.score(X_tr, y_tr)
```

```python
y_pr = model.predict(X_te)
```

```python
accuracy_score(y_te, y_pr)
```

```python
k = 20
```

```python
k_folds = KFold(n_splits=k, shuffle=True)
```

```python
cvscore = cross_val_score(estimator=SVC(kernel="linear"), n_jobs=-1, 
                          X=X_tr, y=y_tr, cv=k_folds, verbose=True, scoring="accuracy")
```

```python
cvscore
```

```python
np.mean(cvscore)  # Average Accuracy
```

```python
sns.heatmap(confusion_matrix(y_te, y_pr), annot=True, fmt="d")
```

```python
print(classification_report(y_te, y_pr))
```
