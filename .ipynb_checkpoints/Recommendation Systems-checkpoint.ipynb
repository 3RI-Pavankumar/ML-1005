```python
df = pd.merge(user, movie, on="item_id")
```

```python
df.head()
```

```python
df.drop(columns="timestamp", inplace=True)
```

```python
df.head()
```

```python
df["title"].value_counts().head(10)
```

```python
df["title"].value_counts().tail(10)
```

```python
movie_counts = df["title"].value_counts().reset_index()
```

```python
movie_counts[movie_counts["count"]<50].shape
```

```python
movie_counts.info()
```

```python
movie_greater_50 = movie_counts[movie_counts["count"]>=50]["title"]
```

```python
movie_greater_50
```

```python
df = df[df["title"].isin(movie_greater_50)]
```

```python
df["title"].value_counts()
```

```python
df.groupby("title")["rating"].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby("title")["rating"].mean().sort_values().head(10)
```

```python
movie_matrix = pd.pivot_table(df, index="user_id", columns="title", values="rating")
```

```python
movie_matrix.head()
```

```python
movie_corr = movie_matrix.corr()
```

```python
movie_corr
```

```python
movie_corr["Star Wars (1977)"].sort_values(ascending=False)[1:11]
```

```python
movie_corr["Absolute Power (1997)"].sort_values(ascending=False)[1:11]
```
