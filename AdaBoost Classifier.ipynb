```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings("ignore")
```

```python
X, y = make_classification(n_samples=1000, n_classes=4, n_informative=5)
```

```python
X.shape
```

```python
y.shape
```

```python
y
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, train_size=0.7, random_state=34)
```

```python
base_model = DecisionTreeClassifier(max_depth=1)
```

```python
model = AdaBoostClassifier(estimator=base_model, n_estimators=200)
```

```python
model.fit(X_tr, y_tr)
```

```python
y_pr = model.predict(X_te)
```

```python
accuracy_score(y_te, y_pr)
```
