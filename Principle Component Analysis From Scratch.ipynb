```python
import numpy as np
```

```python
np.random.seed(0)
X = np.random.rand(100, 2)
```

```python
X
```

```python
mean = np.mean(X, axis=0)
```

```python
mean
```

```python
X_centered = X - mean
```

```python
X_centered
```

```python
cov_matrix = np.cov(X_centered, rowvar=False)
```

```python
cov_matrix
```

```python
eig_val, eig_vec = np.linalg.eig(cov_matrix)
```

```python
eig_val
```

```python
eig_vec
```

```python
sorted_index = np.argsort(eig_val)[::-1]
```

```python
sorted_index
```

```python
eig_val = eig_val[sorted_index]
```

```python
eig_val
```

```python
eig_vec = eig_vec[:, sorted_index]
```

```python
eig_vec
```

```python
k = 1
```

```python
top_k_eig_vec = eig_vec[:,:k]
```

```python
top_k_eig_vec
```

```python
pc = np.dot(X_centered, top_k_eig_vec)
```

```python
pc
```

```python
np.round(pc, decimals=3)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=1)
```

```python
pc_sklearn = pca.fit_transform(X)
```

```python
pc_sklearn
```

```python
np.round(pc_sklearn, decimals=3)
```

```python
np.allclose(pc, pc_sklearn)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```
