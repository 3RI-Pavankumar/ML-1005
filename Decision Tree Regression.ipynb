```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
data = {
    "Area" : [1200, 1500, 1800, 1000, 2000, 1600, 1400, 1700, 1300, 1100],
    "Bedrooms" : [2, 3, 4, 2, 3, 3, 2, 3, 2, 2],
    "Age" : [5, 10, 2, 8, 3, 6, 4, 9, 7, 1],
    "Price" : [300000, 400000, 500000, 250000, 600000, 450000, 350000, 480000, 320000, 280000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
X = df.drop("Price", axis=1)
y = df["Price"]
```

```python
X.head()
```

```python
y.head()
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, train_size=0.7, random_state=342)
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(X_tr, y_tr)
```

```python
y_pr = model.predict(X_te)
```

```python
mse = mean_squared_error(y_te, y_pr)
```

```python
mse
```

```python
r2_score(y_te, y_pr)
```
